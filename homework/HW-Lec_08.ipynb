```python
import matplotlib.pyplot as plt
from random import *
import numpy as np
import mpld3

x = range(0, 1000)
y = []

for N in x:                 #assigns a random assortment of integers from 0 to 50 into the y named array
    y.append(randint(0,50))

graph1 = np.polyfit(x, y, 1 , full = True) #Graphs the first polynomial, which in this case is linear
y1 = np.polyval(graph1[0],x)

graph2 = np.polyfit(x, y, 2 , full = True) #Graphs the second polynomial, and so ony2 = np.polyval(graph2[0],x)

graph3 = np.polyfit(x, y, 3 , full = True)
y3 = np.polyval(graph3[0],x)

graph4 = np.polyfit(x, y, 4 , full = True)
y4 = np.polyval(graph4[0],x)

graph5 = np.polyfit(x, y, 5 , full = True)
y5 = np.polyval(graph5[0],x)




#The following outputs the graphs of all five polynomials fitted against te randomly generated numbers, and gives a residual value on said graph
plt.figure(figsize = (10,10))
plt.plot(x,y,"bo", label="Random Numbers")
plt.plot(x, y1, "c", label="First Degree(Linear)")
plt.plot(x,y2, "k", label = "Second Degree")
plt.plot(x,y3, "r", label = "Third Degree")
plt.plot(x, y4, "y", label = "Fourth Degree")
plt.plot(x, y5, "g", label = "Fifth Degree")
plt.legend()
plt.show()


print("The residuals for graph 1 is", graph1[1])
print("The residuals for graph 2 is", graph2[1])
print("The residuals for graph 3 is", graph3[1])
print("The residuals for graph 4 is", graph4[1])
print("The residuals for graph 5 is", graph5[1])

print("The graph with the lowest residual value is the most accurate fit of the random values, which is the Fifth degree    graph")
```
