```python
from numpy import linspace, zeros, absolute
from matplotlib import pyplot as plt

x_min, x_max, increments1 = -2.0, 2.0, 1000 #scale for the x-axis, ranging from -2 to 2
y_min, y_max, increments2 = -2.0, 2.0, 1000 #scale for y-axis, ranging from -2 to 2
max_Iterations = 100

x_space = linspace(x_min, x_max, increments1)
y_space = linspace(y_min, y_max, increments2)

mand_Array = zeros([increments1,increments2], float)#sets up a 1000x1000 array using float values, filled with zeros

for i in range(len(y_space)): 
    for j in range(len(x_space)):
        z = 0        
        c = x_space[j] + y_space[i]*1j #equation for c, c = x+iy       
      
        
        for k in range(max_Iterations):
             z = z**2+c # provides the next value for z, using its previous value and the value of c calculated above
             if absolute(z)>2: 
                mand_Array[i,j]=k
                break
            


fig = plt.figure(figsize=(10, 10), dpi=150)              
ax = fig.add_axes([0.0, 0.0, 1.0, 1.0], frameon=False, aspect=1)
plt.imshow(mand_Array,origin="lower",extent=[x_min,x_max,y_min,y_max])
plt.show()
```
