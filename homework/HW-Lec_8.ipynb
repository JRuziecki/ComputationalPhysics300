```python
import matplotlib.pyplot as plot
import numpy as np
from math import pi

plot.figure(figsize = [10,6])

f = lambda x: 16668 * (np.sin(x-.0548)-np.sin(0.0548))
a,b = -1.1,1.1
npoints = 20
order = 5


np.random.seed(0)
x = np.linspace(a,b,npoints)
y = f(x) + np.random.rand(npoints) - 0.5

fit = np.polyfit(x, y, order, full=True)
y_p = np.polyval(fit[0], x)

plot.plot(x,f(x), label='ideal function')
plot.plot(x,y, 'o', label='noisy data')
plot.plot(x,y_p, label='polyfit'+str(order))

plot.xlabel('x')
plot.ylabel('$f(x)$')
plot.xlim([a,b])
plot.legend(fontsize=12)
plot.title('fitted coefficients' + str(fit[0]) + '\n residual: ' + str(fit[1]))
plot.show()
```
