```python
from scipy.integrate import quad 
from numpy import exp 
import matplotlib.pyplot as plt


rho = 6.022 * (10 ** 28)   #Density value
DTemp = 428   #Given Debyre Temp. in Kelvin            
V = 1000 * (10**-6)   #Volum of the solid      
k = 1.381 * (10**-23)  #Blotzmann constant     



f = lambda x: ((x**4)*(exp(x)))/((exp(x)-1)**2)  #Calculates the value of the integral to be used in the final equation


#Calculates the heat capacity 
C = lambda T: 9 * V * rho * k * ((T/DebTemp) ** 3) * quad(f, 0 , DebTemp/T)[0]


#Prompts user to input some temperature to be used in the heat capacity calculation
heatcap = float(input("Please enter a temperature in Kelvin: "))
print("The heat capacity for temperature {:.2f} K is {:.2f} J/K.".format(heatcap, C(heatcap)))

```

```python
#A graph of heat capacity as a function of temperature
Temps = range(5, 500, 1)
HeatCaps = []
for t in Temps:
    HeatCaps.append(C(t))

plt.figure(figsize=(10,10))
plt.plot(Temps,HeatCaps)
plt.xlabel("Temperature(K)", size = 30)
plt.ylabel("$C_V$ (J/K)", size = 30)
plt.title("Heat Capacity vs Temperature", size = 30)
plt.show()
```
